```python
from pandas import ExcelWriter
from pandas import ExcelFile
sub = pd.read_excel('ME09_ListOfSuburbs.xlsx')
mon = pd.read_excel('Monuments.xlsx')
#mon.head()
#mon['sub_id']
```

```python
mon.head()
```

```python
# for m in mon.values:
sub_id = []
for i in range(len(mon.values)):
    for s in sub.values:
        if (mon.values[i][5].lower().strip() == s[1].lower().strip()) & (mon.values[i][6] == s[2]):
            sub_id.append(s[0])
            print(s[0])
            continue
```

```python
mon['sub_id'] = sub_id
```
