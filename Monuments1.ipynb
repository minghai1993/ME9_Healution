```python
import pandas as pd
```

```python
places1 = pd.read_csv('Melbourne_public_artwork.csv')
places1.head()
```

```python
places1["is_duplicate"]= places1.duplicated()
places1
```

```python
places1.isnull().sum()
```

```python
places1 = places1.drop(["EASTING","NORTHING","ALTNAME","RESPAUTHOR","MELWAYREF","XORG","XSOURCE","is_duplicate"], axis = 1)
places1.head()
```

```python
places1[['Latitude','Longitude']] = places1.geom.str.split(",",expand=True,)
places1.head()
```

```python
places1 = places1.drop(["geom"], axis = 1)
```

```python
places1 = places1[['Latitude', 'Longitude', 'ASSET_TYP', 'NAME', 'ADDRESSPT', 'STRUCTURE_', 'ARTDATE', 'ARTIST']]
places1.head()
```

```python
places1.columns = ['Latitude', 'Longitude', 'Asset_type', 'Name', 'Address', 'Structure', 'Artdate', 'Artist']
places1.head()
```

```python
places1.to_csv('./Mm2.csv', sep='\t', encoding='utf-32')
```
