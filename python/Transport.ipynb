```python
import pandas as pd
```

## Railway Station

```python
a1 = pd.read_csv('stops.txt')
a1.head()
```

```python
a1["is_duplicate"]= a1.duplicated()
```

```python
a1.columns = ['Stop_id', 'Stop_name', 'Latitude', 'Longitude','is_duplicate']
a1.head()
```

```python
a1['Stop_name'] = a1['Stop_name'].str.replace('(','_').str.replace(')','')
```

```python
a1[['Desription','Suburb']] = a1['Stop_name'].str.split('_', n =1, expand = True)
```

```python
a1 = a1.drop(["is_duplicate","Stop_name"], axis = 1)
```

```python
a1.head()
```
